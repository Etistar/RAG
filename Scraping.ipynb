```python
import requests
from bs4 import BeautifulSoup

# Liste des pages à scraper
pages = [
  "https://impactlanguagementors.com/fr/",
    "https://impactlanguagementors.com/fr/about-us/",
    "https://impactlanguagementors.com/fr/facilities/",
    "https://impactlanguagementors.com/fr/admission/",
    "https://impactlanguagementors.com/fr/contact-us/"
]

headers = {
    "User-Agent": "Mozilla/5.0"
}

def extract_structured_text_between_keywords(soup, start_keyword="Apply Now", end_keyword="Main Menu"):
    # Extraire tout le texte brut
    full_text = soup.get_text(separator="\n", strip=True)

    # Repérer les positions de début et de fin
    start = full_text.rfind(start_keyword)
    end = full_text.find(end_keyword, start)
    
    if start == -1 or end == -1 or start >= end:
        return []

    # Extraire le texte entre les deux
    segment_text = full_text[start + len(start_keyword):end].strip()

    # Créer une soupe réduite à ce segment
    reduced_soup = BeautifulSoup(segment_text, "html.parser")

    # Alternative : travailler sur le contenu brut en découpant ligne par ligne
    lines = segment_text.splitlines()
    structured = []
    current_title = None
    current_content = []

    for line in lines:
        line = line.strip()
        if not line:
            continue
        # Un titre si tout en majuscules ou marqué en gras peut être détecté par règle simple ici
        if line.isupper() or len(line) < 20:
            if current_title:
                structured.append((current_title, '\n'.join(current_content)))
            current_title = line
            current_content = []
        else:
            if current_title:
                current_content.append(line)

    if current_title and current_content:
        structured.append((current_title, '\n'.join(current_content)))

    return structured

# Contenu global à sauvegarder
full_output = ""
# Traitement de chaque page
for url in pages:
    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, "html.parser")

        sections = extract_structured_text_between_keywords(soup)

        full_output += f"\n--- Contenu structuré extrait de : {url} ---\n"
        for title, content in sections:
            section_text = f"\n {title}\n{content}\n"
            full_output += section_text
            
    except requests.RequestException as e:
        print(f"Erreur lors de la requête vers {url} : {e}")
# Sauvegarde dans un seul fichier
with open("output_impact_all.txt", "w", encoding="utf-8") as f:
    f.write(full_output)

print("Tous les résultats ont été enregistrés dans output_impact_all.txt")
```
